### Understanding Marketing Funnels

The whole team had one problem in understanding the layers and how they related to eachother. To understand this, we had to research a little bit more into the **marketing funnel**. 

A marketing funnel is a process turning _leads into customers_ by slowly nurturing prospective customers through the purchasing decision, narrowing down customers in each stage of funnel. 






            
1. **Awareness:** - Potential customers are drawn into this stage through marketing campaigns. Here, _lead generation_ takes place where information is collected. 

2. **Interest & Consideration:** - Once leads are generated, they _move_ onto the interest stage where they **_learn more about the company's products for e.g._**. Marketers can nurture leads through emails, content that is more targeted around the brand. Leads have been changed into **marketing qualified leads** and are seen as prospective customers. Marketers can send prospects more information about products and offers through automated email campaigns

4. **Intent:** To get to the intent stage, prospects must demonstrate that they are interested in buying the product. This can happen in a survey, or when a product is placed in the shopping cart on an ecommerce website.

5. **Evaluation:** Buyers make final decision about whether or not to buy product. Typically, marketing and sales work together to closely nurture the decision making process and convince the buyer that their brand's product is the best choice. 

6. **Acquisition:** A positive experience on the part of the buyer can lead to referrals that fuel the top of the marketing funnel.
